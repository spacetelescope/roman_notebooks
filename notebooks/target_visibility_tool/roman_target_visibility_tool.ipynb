## Introduction

The Roman Target Visibility Tool (RTVT) is an interactive planning utility for estimating when astronomical targets are observable by the Nancy Grace Roman Space Telescope. RTVT evaluates target visibility using Roman's Sun-angle observing constraints, then summarizes observable windows over a user-selected time interval. For more info on Roman's coordinate systems and the parameters used here for calcualting the visibility, please see [Roman Coordiante Systems](https://roman-docs.stsci.edu/data-handbook/wfi-data-levels-and-products/coordinate-systems) and [Roman Orbit Visualization](https://science.nasa.gov/mission/roman-space-telescope/observatory-technical/).

This notebook launches the interactive RTVT sky-map and Gantt viewer. It is designed for exploratory scheduling checks, quick comparisons among multiple targets, and visual demonstrations of Roman target visibility behavior.

The viewer allows users to:

- Select targets directly from an all-sky visibility map.
- Search for named targets such as `Andromeda`, `M31`, or `Vega`.
- Enter exact coordinates when a precise position is preferred.
- Switch among equatorial, Galactic, and ecliptic coordinate displays.
- Inspect the latest target's visibility window and Sun-target separation curve.
- Compare all selected targets in a cumulative visibility-window Gantt chart.

## Imports and Viewer Setup

The cell below enables the interactive Matplotlib widget backend and launches the RTVT notebook viewer. The default configuration samples the 2027 observing year once per day, from 2027-01-01 through 2027-12-31, and uses a 10 degree all-sky grid for the overview map.


```python
%matplotlib widget
from importlib import reload
import rtvt_standalone

rtvt_standalone = reload(rtvt_standalone)

viewer = rtvt_standalone.launch_interactive_sky_gantt_with_controls(
    grid_step_deg=10,
    start_time="2027-01-01",
    duration_days=365,
    sampling_days=1,
)
```

## Using the Interactive Viewer

After running the cell above, use the coordinate-system control to choose equatorial coordinates (RA/Dec), Galactic coordinates (l/b), or ecliptic coordinates (longitude/latitude). Click on the all-sky map to add a target, search by target name, or use the coordinate entry controls for an exact position. Each selected target is assigned a consistent color that is reused in the map marker, visibility-window plot, Gantt row, and Sun-target separation comparison.

The most recently selected target is highlighted in the Gantt chart so that its row is easy to identify when several targets have been added.


## Interpreting the Outputs

The all-sky map shows the fraction of the sampled year for which each sky position satisfies Roman's target visibility constraints. Higher values indicate regions with longer visibility over the selected interval.

The latest-target visibility plot shows the time ranges when the selected target is observable, together with its Sun-target separation. The allowed observing zone is defined by Roman's Sun-angle limits, so visibility changes as the apparent Sun position moves through the year.

The selected-target Gantt chart summarizes each target's observable windows as horizontal bars. This view is useful for comparing many targets at once and identifying overlapping scheduling opportunities.

The Sun-target separation comparison plot overlays the separation curves for all selected targets. It provides a compact way to see why different targets enter or leave the allowed visibility zone at different times.


## Output Tables

The notebook keeps the live display focused on the map and plots. Tables are written when you click **Create report + CSV** rather than being rendered inline after every target selection. The generated CSV folder includes `run_parameters.csv`, `selected_targets.csv`, `observable_windows.csv`, and `sampled_visibility.csv`.

The **Run Parameters** table records the calculation configuration, including the display coordinate system, interval start and end, duration, cadence, sample count, all-sky grid step, allowed Sun-target separation range, and CVZ threshold. The **Selected Target Summary** table gives each target's input/display coordinates, ICRS RA/Dec, Galactic coordinates, ecliptic coordinates, visible samples, visible days, visibility fraction, observable-window count, first and last visible dates, Sun-target separation range, nominal-roll range, and CVZ flag. The **Observable Windows** table lists each contiguous visibility window with start/end dates, duration, nominal-roll endpoints, and Sun-target separation endpoints. The **Sampled Visibility** CSV contains every sampled row and every per-sample quantity produced by the visibility calculator: Sun RA/Dec, Sun-target separation, visibility flag, nominal roll, observatory/FPA position angles, and Sun angles in the observatory frame.
